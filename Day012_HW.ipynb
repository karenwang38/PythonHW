### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？
Ans: 用dict分類別去存標題。

```python
news_category = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    # 分類
    category_temp = d.find('em').text
    # 內容
    category_context = d.find(class_="date").text + ' ' + d.find_all('a')[-1].text
    try:
        list(news_category.keys()).index(category_temp)
#         print('category is exist!')
        news_category[category_temp].append(category_context)
    except:    
#         print('category Not exist, add it.')
        news_category.setdefault(category_temp, [category_context])
news_category
```

### ③ 哪一個類別的文章最多

```python
for i in news_category:
    print(len(i))
```
