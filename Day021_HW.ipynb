## 1. 取出今天所有的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver')

# 取得資料，丟到 BeautifulSoup 解析
browser.get("https://www.ettoday.net/news/news-list.htm")


```

```python
import time
import datetime
beforeDays = 0
now = datetime.datetime.now()
temp = str(now).split(' ')[0].split('-')
print(temp)
today_data = '/'.join(temp)



html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

while soup.find(class_="part_list_2").find_all('h3')[-1].text.split(' ')[0] == today_data:
    print(soup.find(class_="part_list_2").find_all('h3')[-1].text.split(' ')[0], soup.find(class_="part_list_2").find_all('h3')[-1].text.split(' ')[1])
    # 每個兩秒鐘自動往下滑
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000000);")
    print('模擬滑鼠下滑！')
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")

print('today news donwload done!')       
```

```python
# 印出今天所有的文章標題
news = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text.split(' ')[0] != today_data:
        break
    else:
#         print(d.find(class_="date").text, d.find_all('a')[-1].text)
        news.append({
            '時間': d.find(class_="date").text,
            '標題': d.find_all('a')[-1].text
        })

import pandas as pd
pd.DataFrame(news)
```

```python
browser.close()
```
