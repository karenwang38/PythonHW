### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find_all(class_="r-ent"):
    print('====================')
    print('(首頁)標題:', d.find(class_='title').text.replace('\t', '').replace('\n', ''))
    print('(首頁)作者:', d.find(class_='author').text)
    print('(首頁)時間:', d.find(class_='date').text)
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")  
        for i in r.find_all(class_='article-metaline'):     
            print(i.find(class_='article-meta-tag').text, i.find(class_='article-meta-value').text)
    except:
        continue
```

### ③ 試著爬爬看其他版的文章
加爬內容與推文

```python
url = 'https://www.ptt.cc/bbs/marriage/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find_all(class_="r-ent"):
    print('====================')
#     print('(首頁)標題:', d.find(class_='title').text.replace('\t', '').replace('\n', ''))
#     print('(首頁)作者:', d.find(class_='author').text)
#     print('(首頁)時間:', d.find(class_='date').text)
    
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        
        for i in r.find_all(class_='article-metaline'):     
            print(i.find(class_='article-meta-tag').text, i.find(class_='article-meta-value').text)
        bbs_context = r.find(class_='bbs-screen bbs-content')
        print('\n文章內容：\n', bbs_context.text.split('--')[0]) 
        print('推文=======：', bbs_context.text.split('--')[1])
        print('\n\n\n')    
    except:
        continue
```
