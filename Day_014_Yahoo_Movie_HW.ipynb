# YAHOO電影爬蟲練習
## 練習爬取電影放映資訊。必須逐步獲取電影的代號、放映地區、放映日期後，再送出查詢給伺服器。

```python
import requests
import re
from bs4 import BeautifulSoup
```

### 先搜尋全部的電影代號(ID)資訊

```python
# 查看目前上映那些電影，並擷取出其ID資訊
url = 'https://movies.yahoo.com.tw/'
resp = requests.get(url)
resp.encoding = 'utf-8'

soup = BeautifulSoup(resp.text, 'lxml')
html = soup.find("select", attrs={'name':'movie_id'})
movie_item = html.find_all("option", attrs={'data-name':re.compile('.*')})

for p in movie_item:
    print("Movie: %s, ID: %s" % (p["data-name"], p["value"]))
```

```python
# 從網頁爬電影基本資料
import urllib.parse

'''
https://movies.yahoo.com.tw/movieinfo_main/%E7%BD%AA%E6%A8%82%E5%9C%92-the-promised-land-10349
https://movies.yahoo.com.tw/movieinfo_main/-the-promised-land-10349
中間不需要電影名稱，只要電影id即可顯示該電影資料
'''
move_name = '罪樂園'
urlUTF8encodeStr = urllib.parse.quote(move_name)
moveInfo_url = 'https://movies.yahoo.com.tw/movieinfo_main/'+ urlUTF8encodeStr + '-the-promised-land-' + str(movie_id)
# https://movies.yahoo.com.tw/movieinfo_main/%E7%BD%AA%E6%A8%82%E5%9C%92-the-promised-land-10349
# print(urllib.parse.quote("罪樂園"))  
# %E7%BD%AA%E6%A8%82%E5%9C%92
print(moveInfo_url)
r = requests.get(moveInfo_url)
# print(r.text)
soup = BeautifulSoup(r.text, 'lxml')
movie_intro_info_r = soup.find(class_='movie_intro_info_r')
# print(movie_intro_info_r)
print('電影名稱：', movie_intro_info_r.find('h1').text)

# 演員與導演資料整理
movie_intro_list = []
for j in movie_intro_info_r.find_all(class_='movie_intro_list'):
    movie_intro_list.append(j.text.replace('\n', '').replace(' ', ''))
#     print(movie_intro_list)

actor_cnt=0
for i in movie_intro_info_r.find_all('span'):
    
    if (i.text == '導演：') | (i.text == '演員：'):
        print(i.text, movie_intro_list[actor_cnt])
        actor_cnt += 1
    elif (i.text == '官方連結：'):
        print(i.text, movie_intro_info_r.find('a')['href'])
    else:
        print(i.text)
```

### 指定你想要觀看的放映地區，查詢有上映電影的場次日期

```python
# 指定放映地區
area_id = 28
```

```python
# 向網站發送請求
url = 'https://movies.yahoo.com.tw/movietime_result.html'

payload = {'movie_id':str(movie_id), 'area_id':str(area_id)}
resp = requests.get(url, params=payload)
resp.encoding = 'utf-8'

soup = BeautifulSoup(resp.text, 'lxml')
movie_date = soup.find_all("label", attrs={'for':re.compile("date_[\d]")})

print(url)
# 列印播放日期
for date in movie_date:
    print("%s %s" % (date.p.string, date.h3.string))
```

### 最後指定觀看的日期，查詢並列印出放映的電影院、放映類型(數位、3D、IMAX 3D...)、放映時間等資訊。

```python
# 選定要觀看的日期
date = "2020-03-12"
```

```python
# 向網站發送請求，獲取上映的電影院及時間資訊
url = "https://movies.yahoo.com.tw/ajax/pc/get_schedule_by_movie"
payload = {'movie_id':str(movie_id),
           'date':date,
           'area_id':str(area_id),
           'theater_id':'',
           'datetime':'',
           'movie_type_id':''}

resp = requests.get(url, params=payload)
# print(resp.json()['view'])  # 若有需要，列印出json原始碼

soup = BeautifulSoup(resp.json()['view'], 'lxml')
html = soup.find_all("ul", attrs={'data-theater_name':re.compile(".*")})

for i in html:
    print(i)
    print('========\n\n\n')
```
